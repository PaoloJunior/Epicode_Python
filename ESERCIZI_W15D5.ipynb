```python
import dotenv
import os
from sqlalchemy import create_engine
import pandas as pd
dotenv.load_dotenv(dotenv_path=".env", override=True)
username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")
conn_string = "mysql+pymysql://" + username + ":" + password + "@" + host + "/" + dbname
db_engine = create_engine(conn_string)
query= "SELECT * FROM dimsalesterritory"
df3=pd.read_sql(query,db_engine)
```

```python
df_merged = pd.merge(df1, df2, left_on='EmployeeKey', right_on='EmployeeKey', how='inner')
print("DataFrame risultante dopo il join tra df1 e df2:")
print(df_merged.head())
print(f"Dimensione del DataFrame risultante: {df_merged.shape}")
```

```python

df_final = pd.merge(df_merged, df3, left_on='SalesTerritoryKey', right_on='SalesTerritoryKey', how='inner')


count_by_country_region = df_final.groupby(['SalesTerritoryCountry', 'SalesTerritoryRegion']).size()


print("Conteggio dei dipendenti per ogni paese e regione:")
print(count_by_country_region)

mean_baserate_by_country = df_final.groupby('SalesTerritoryCountry')['BaseRate'].mean()

print("Media del BaseRate per ogni paese:")
print(mean_baserate_by_country)
```
